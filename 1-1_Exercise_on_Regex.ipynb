```python
def f(x):
    return x.groups()[0][0:3]
```

```python
sol0 = df['text'].str.extract('(\d?\d:(\d\d) ?([ap]m))')
sol0
```

```python
sol = df['text'].str.extractall('(\d?\d:(\d\d) ?([ap]m))')
sol
```

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
meals = ['breakfast', 'lunch', 'dinner']
```
