### Compute the probability that randomly selected country with Happiness score over 6.0 is from Western Europe. You will have to use pandas to count the appropriate quantities.


I assume the follwitng events:
   + A: select randomly a country for Western Europe and 
   + B: select randomly a country with Happiness Score > 6.0
   
   We want to find the conditional probability P(A/B) where according to Bayesian P(A/B) = P(A∪B)/P(B).


```python
df
```

```python
df.reset_index(inplace = True)
```

```python

df.loc[df['Happiness Score'] > 6, :].groupby('Region')['Country'].count().to_frame()
```

```python
df.groupby('Region')['Country'].size().to_frame().shape[1]
```

```python
# calculation of P(B), Pb
Total = len(df)

(df['Happiness Score'] > 6.0).sum()
Pb = (df['Happiness Score'] > 6.0).sum()/Total
```

```python
# calculation of P(A∪B), Pab
Pab = ((df['Happiness Score'] > 6) & (df['Region'] == 'Western Europe')).sum()/Total
```

```python
P = Pab/Pb
```

```python
print( "the probabilita a randomly selected country with Happiness Score over 6 is from Western Europe is {}".format(Pab/Pb))
```

# H. Matrices
### Define a matrix whose rows correspond to countries and the columns to the regions. Fill in the matrix with 0/1 values where entry (i,j) is a 1 if the country in row i is in the region in column j and a 0 otherwise.

I tryied out two solutions for this task: one without Broadcasting and one applying Broadcasting

solution without Broadcasting:

```python
df.set_index('Country')
```

```python
df
```

```python
df.Region.unique()
```

```python
first_time = 0
for region in df.Region.unique():
    temp = df[df['Region'] == region].Region.copy().to_frame()
    temp['new'] = temp['Region'].map({region: int(1)}).copy()
    temp.drop('Region', axis = 1, inplace = True)
    temp.rename(columns ={'new':region}, inplace = True)
    first_time += 1
    if first_time == 1: 
        s = temp.copy()
    else:
        s = pd.concat([s,temp],axis=1, sort=False)
```

```python
s.fillna(0, inplace = True)
```

```python
s
```

```python
result = s.to_numpy().astype(int).copy()
```

```python
result
```
